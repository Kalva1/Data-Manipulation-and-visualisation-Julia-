```julia
using StatsPlots
a=@df df plot(:Episode_number, [:Votes],color=[:red],label="Episode ratings with votes",xlabel="episodenumber",ylabel="votes")
b=@df df plot(:Episode_number, [:Stars],color=[:green],label="Episode ratings with Stars",xlabel="episodenumber",ylabel="stars")
plot(a,b)
```

```julia
scatter(df.Season,df.Votes)
```

```julia
scatter(df.Season,df.Stars)
```

```julia
a=@df df bar(:Season, [:Votes],color=[:red],label="Season ratings with Votes")
b=@df df bar(:Season, [:Stars],color=[:green],label="Season ratings with Stars")
plot(a,b)
```

```julia
using StatsPlots
@df df scatter(:Stars,:Votes,
    group = :Season,
    m = (0.9, [:+ :h :star7], 15),
    bg = RGB(0.2, 0.2, 0.2),
    xlabel="stars",
    ylabel="votes"
)
```

```julia
p1=plot(df.Season,df.Votes)
p2=scatter(df.Season,df.Votes)
p3=bar(df.Season,df.Votes)
p4=histogram(df.Votes)
plot(p1,p2,p3,p4,layout=(2,2),legend=false)
```
